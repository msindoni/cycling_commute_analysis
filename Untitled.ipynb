# Looking at relationship between speed and temperature

```python
data=data.loc[(data['incident?']=='n') & pd.to_numeric(data['back_temp'], errors='coerce').notnull()] #removing values that are not numeric

# Convert 'there_temp' column to numeric
data['back_temp'] = pd.to_numeric(data['back_temp'])

plt.scatter(data['back_temp'], data['back_total_time_sec'])
# plt.ylim(0,)

# Compute correlation coefficient
correlation_coefficient = data['back_temp'].corr(data['back_total_time_sec'])

print("Correlation Coefficient:", correlation_coefficient)


x = np.array([min(data['back_temp']), max(data['back_temp'])])
y = correlation_coefficient * x  # y = mx
plt.plot(x, y, color='red', label=f'Correlation Line (r={correlation_coefficient:.2f})')


plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.regplot(data=data, x='back_temp', y='back_total_time_sec', scatter_kws={'s': 100})  # Increase marker size
plt.xlabel('Back Temperature')
plt.ylabel('Back Total Time (sec)')
plt.title('Scatter Plot with Regression Line')
plt.grid(True)
plt.show()


from scipy.stats import pearsonr
# Compute Pearson correlation coefficient and p-value
correlation_coefficient, p_value = pearsonr(data['back_temp'], data['back_total_time_sec'])

print("Correlation Coefficient:", correlation_coefficient)
print("p-value:", p_value)
```

# Generating averages for all time and past 10 rides

```python
data=df_master.loc[(df_master['back_total_time_sec']>800) & (df_master['back_total_time_sec']<1500)]


column_name='back_total_time_sec'
all_time_avg_list=[]
ten_day_average_list=[]
ride_count_list=[]

ride_count=1
total_sec_sum=0

for i in range(len(data)):

    #generating all time average
    total_sec_sum+=data[column_name].iloc[i]
    all_time_avg_list.append(total_sec_sum/ride_count)
    
    #generating ten day average
    if i<10: #first 10 will avaerage one at a time, not all 10 at once, same function as above, different list
        ten_day_average_list.append(total_sec_sum/ride_count)
    else:
        ten_day_average=(data[column_name].loc[i-9:i].sum())/10 #since indexing withy pandas is inclusive
        ten_day_average_list.append(ten_day_average)
    
    ride_count_list.append(ride_count)   
        
    ride_count+=1
```

```python
plt.plot(ride_count_list, all_time_avg_list)
```

```python
plt.plot(ride_count_list, ten_day_average_list)
```
